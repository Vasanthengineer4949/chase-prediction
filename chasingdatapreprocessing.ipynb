```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')
```

```python
deliveries.groupby(['match_id', 'inning']).sum()['total_runs']
```

```python
inn_score = deliveries.groupby(['match_id', 'inning']).sum()['total_runs'].reset_index()
inn_score = inn_score[inn_score['inning'] == 1]
inn_score
```

```python
matches = matches.merge(inn_score[['match_id','total_runs']], left_on='id', right_on='match_id')
matches
```

```python
matches['team1'] = matches['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
matches['team2'] = matches['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

matches['team1'] = matches['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
matches['team2'] = matches['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
teams = [
    'Chennai Super Kings',
    'Delhi Capitals',
    'Mumbai Indians',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Rajasthan Royals',
    'Royal Challengers Bangalore',
    'Sunrisers Hyderabad'
]
```

```python
matches = matches[matches['team1'].isin(teams)]
matches = matches[matches['team2'].isin(teams)]
```

```python
matches =matches[matches['dl_applied'] == 0]
```

```python
matches
```

```python
matches = matches[['match_id','city','winner','total_runs']]
deliveries = matches.merge(deliveries, on="match_id")
deliveries = deliveries[deliveries['inning'] == 2]
```

```python
deliveries
```

```python
deliveries['current_score'] = deliveries.groupby('match_id').cumsum()['total_runs_y']
deliveries
```

```python
deliveries['runs_left'] = deliveries['total_runs_x'] - deliveries['current_score']
deliveries['balls_left'] = 120 - ((deliveries['over']-1)*6 + deliveries['ball'])
deliveries['player_dismissed'] = deliveries['player_dismissed'].fillna("0")
deliveries['player_dismissed'] = deliveries['player_dismissed'].apply(lambda x:x if x == "0" else "1")
deliveries['player_dismissed'] = deliveries['player_dismissed'].astype('int')
wkts = deliveries.groupby(['match_id']).cumsum()['player_dismissed'].values
deliveries['wickets'] = 10 - wkts
```

```python
deliveries
```

```python
deliveries['current_rr'] = (deliveries['current_score']*6) / (120 - deliveries['balls_left'])
deliveries['required_rr'] = (deliveries['runs_left']*6) / deliveries['balls_left']
```

```python
deliveries
```

```python
def match_result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
deliveries['result'] = deliveries.apply(match_result ,axis=1)
```

```python
deliveries
```

```python
deliveries.columns
```

```python
final_df = deliveries[['batting_team','bowling_team','city','runs_left','over', \
                    'balls_left','wickets','total_runs_x','current_rr','required_rr','result']]
```

```python
final_df.dropna(inplace=True)
final_df = final_df[final_df['runs_left'] > 0]
final_df = final_df[final_df['balls_left'] != 0]
```

```python
final_df.to_csv('chasing.csv', index=False)
```
